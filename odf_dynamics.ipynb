```python
import matplotlib.pyplot as plt
import numpy as np
from odf import ODF

odf = [] 
for RA in range(0, 100, 5):
    row = []
    for MDF in range(0, 100, 5):
        val = ODF(MDF/100, RA/100)
        row.append(val)
    odf.append(row)
odf = np.array(odf).reshape(20,20)
odf
print(odf.shape)
```

```python
plt.figure(figsize=(16,12))
ra_range = [r / 100 for r in range(0, 100, 5)]
mdf_range = [r / 100 for r in range(0, 100, 5)]

confMat=odf

heatmap = plt.pcolor(confMat)

for y in range(confMat.shape[0]):
    for x in range(confMat.shape[1]):
        plt.text(x + 0.5, y + 0.5, '%.3f' % confMat[y, x],
                horizontalalignment='center',
                verticalalignment='center',)


#plt.grid() #this will look bad now
plt.colorbar(heatmap)
plt.subplots_adjust(left=0.15, right=0.99, bottom=0.15, top=0.99)
plt.ylabel('RA')
plt.xlabel('MDF')

plt.xticks(np.arange(len(mdf_range))+0.5, mdf_range)
plt.yticks(np.arange(len(ra_range))+0.5, ra_range)
plt.show()
```
